```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def explore_data(file_path, sheet_name):
    # Load the data
    data = pd.read_excel(file_path, sheet_name=sheet_name)

    # 1. Study each attribute and identify the datatype
    print("\nData Types and Summary:")
    data_types = data.dtypes
    print(data_types)

    # Identify categorical and numeric columns
    categorical_cols = data.select_dtypes(include=[object]).columns
    numeric_cols = data.select_dtypes(include=[np.number]).columns

    # Categorical attributes and their values
    print("\nCategorical Attributes and Their Values:")
    for col in categorical_cols:
        unique_values = data[col].unique()
        print(f"{col}: {unique_values}")

    # 2. Encoding schemes
    print("\nEncoding Scheme Recommendations:")
    for col in categorical_cols:
        unique_values = data[col].unique()
        num_unique = len(unique_values)
        print(f"{col} ({'Nominal' if num_unique > 10 else 'Ordinal'}): {unique_values}")

    # 3. Study the data range for numeric variables
    print("\nNumeric Attributes Range:")
    for col in numeric_cols:
        min_val = data[col].min()
        max_val = data[col].max()
        print(f"{col}: Min = {min_val}, Max = {max_val}")

    # 4. Study the presence of missing values
    print("\nMissing Values:")
    print(data.isnull().sum())

    # 5. Study the presence of outliers in data
    print("\nOutliers Detection:")
    for col in numeric_cols:
        plt.figure(figsize=(10, 4))
        sns.boxplot(data[col])
        plt.title(f'Boxplot of {col}')
        plt.show()

    # 6. Calculate mean and variance for numeric variables
    print("\nNumeric Variables Statistics:")
    for col in numeric_cols:
        mean_val = data[col].mean()
        var_val = data[col].var()
        print(f"{col}: Mean = {mean_val}, Variance = {var_val}")

    # Encoding categorical variables based on identified schemes
    print("\nEncoding Categorical Variables:")
    for col in categorical_cols:
        if data[col].dtype == 'object':
            unique_values = data[col].unique()
            num_unique = len(unique_values)
            if num_unique <= 10:  # Assuming nominal if there are few unique values
                # One-Hot Encoding
                encoded_df = pd.get_dummies(data[col], prefix=col, drop_first=True)
                data = pd.concat([data, encoded_df], axis=1)
                data.drop(columns=[col], inplace=True)
                print(f"One-Hot Encoded {col}")
            else:
                # Label Encoding
                data[col] = pd.Categorical(data[col]).codes
                print(f"Label Encoded {col}")

    # Display the cleaned and encoded data
    print("\nEncoded Data:")
    print(data.head())

def main():
    # Path to the Excel file
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'thyroid0387_UCI'

    explore_data(file_path, sheet_name)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Load the data from the Excel file
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'IRCTC Stock Price'
    data = pd.read_excel(file_path, sheet_name=sheet_name)

    print("Initial Data:")
    print(data.head())

    # Impute missing values for numeric columns
    # Check which columns are numeric
    numeric_cols = data.select_dtypes(include=[np.number]).columns

    for col in numeric_cols:
        if data[col].isnull().any():
            if data[col].nunique() / len(data[col]) < 0.1:  # assume  high cardinality is an indicator of outliers
                #  using median for numeric data with outliers
                median_value = data[col].median()
                data[col].fillna(median_value, inplace=True)
            else:
                # Impute using mean for numeric data without outliers
                mean_value = data[col].mean()
                data[col].fillna(mean_value, inplace=True)

    # Impute missing values for categorical columns
    categorical_cols = data.select_dtypes(include=[object]).columns

    for col in categorical_cols:
        if data[col].isnull().any():
            mode_value = data[col].mode()[0]
            data[col].fillna(mode_value, inplace=True)


    print("Cleaned Data:")
    print(data.head())


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

def normalize_data(file_path, sheet_name):
    # Load the data

    data = pd.read_excel(file_path, sheet_name=sheet_name)

    # Identify numeric columns
    numeric_cols = data.select_dtypes(include=[np.number]).columns

    # Study the data range for numeric variables
    print("\nNumeric Attributes Range:")
    for col in numeric_cols:
        min_val = data[col].min()
        max_val = data[col].max()
        print(f"{col}: Min = {min_val}, Max = {max_val}")

    # Identify attributes needing normalization
    print("\nNormalization Needs:")
    normalization_needed = numeric_cols
    print(f"Attributes for normalization: {list(normalization_needed)}")

    # Apply normalization techniques
    # Min-Max Scaling
    minmax_scaler = MinMaxScaler()
    minmax_scaled_data = data.copy()
    minmax_scaled_data[normalization_needed] = minmax_scaler.fit_transform(data[normalization_needed])

    print("\nMin-Max Scaled Data:")
    print(minmax_scaled_data.head())
    # Standardization (Z-score Normalization)
    standard_scaler = StandardScaler()
    standardized_data = data.copy()
    standardized_data[normalization_needed] = standard_scaler.fit_transform(data[normalization_needed])

    print("\nStandardized Data:")
    print(standardized_data.head())

def main():
    # Path to the Excel file
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'IRCTC Stock Price'

    normalize_data(file_path, sheet_name)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

def load_excel_data(file_path, sheet_name):
    return pd.read_excel(file_path, sheet_name=sheet_name)

def filter_binary_data(data):
    # Use map instead of applymap to avoid deprecation warning
    binary_cols = data.columns[data.apply(lambda col: col.map(lambda x: x in [0, 1]).all(), axis=0)]
    binary_data = data[binary_cols]
    return binary_data

def extract_vectors(binary_data):
    if binary_data.shape[0] < 2:
        raise IndexError("The dataset does not contain enough rows for comparison.")

    vector1 = binary_data.iloc[0]
    vector2 = binary_data.iloc[1]

    common_features = vector1.index.intersection(vector2.index)

    return vector1[common_features], vector2[common_features]

def calculate_feature_counts(vector1, vector2):
    f11 = ((vector1 == 1) & (vector2 == 1)).sum()
    f10 = ((vector1 == 1) & (vector2 == 0)).sum()
    f01 = ((vector1 == 0) & (vector2 == 1)).sum()
    f00 = ((vector1 == 0) & (vector2 == 0)).sum()

    return f11, f10, f01, f00

def compute_similarity_measures(f11, f10, f01, f00):
    jc = f11 / (f01 + f10 + f11) if (f01 + f10 + f11) != 0 else 0
    smc = (f11 + f00) / (f00 + f01 + f10 + f11) if (f00 + f01 + f10 + f11) != 0 else 0

    return jc, smc

def main():
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'thyroid0387_UCI'
    data = load_excel_data(file_path, sheet_name)
    binary_data = filter_binary_data(data)

    try:
        vector1, vector2 = extract_vectors(binary_data)
        f11, f10, f01, f00 = calculate_feature_counts(vector1, vector2)
        jc, smc = compute_similarity_measures(f11, f10, f01, f00)
        print(f"Jaccard Coefficient (JC): {jc}")
        print(f"Simple Matching Coefficient (SMC): {smc}")
    except IndexError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder

def load_excel_data(file_path, sheet_name):
    return pd.read_excel(file_path, sheet_name=sheet_name)

def preprocess_data(data):
    # Convert categorical data to numeric
    categorical_cols = data.select_dtypes(include=['object']).columns
    label_encoders = {}

    for col in categorical_cols:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col].astype(str))
        label_encoders[col] = le

    # Fill or drop missing values
    data = data.fillna(0)  # Or use a different strategy if necessary
    return data

def extract_full_vectors(data, row_indices):
    if len(row_indices) < 2:
        raise IndexError("The dataset does not contain enough rows for comparison.")

    vector1 = data.iloc[row_indices[0]].values.reshape(1, -1)
    vector2 = data.iloc[row_indices[1]].values.reshape(1, -1)

    return vector1, vector2

def calculate_cosine_similarity(vector1, vector2):
    cosine_sim = cosine_similarity(vector1, vector2)[0][0]
    return cosine_sim

def main():
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'thyroid0387_UCI'
    data = load_excel_data(file_path, sheet_name)

    # Preprocess the data to handle categorical values and missing values
    data = preprocess_data(data)

    try:
        # Extract full vectors for the first two observations
        vector1, vector2 = extract_full_vectors(data, [0, 1])
        cosine_sim = calculate_cosine_similarity(vector1, vector2)
        print(f"Cosine Similarity: {cosine_sim}")
    except IndexError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

def load_excel_data(file_path, sheet_name):
    return pd.read_excel(file_path, sheet_name=sheet_name)

def filter_binary_data(data):
    binary_cols = data.columns[data.apply(lambda col: col.map(lambda x: x in [0, 1]).all(), axis=0)]
    if len(binary_cols) == 0:
        raise ValueError("No binary columns found in the dataset.")
    return data[binary_cols]

def get_first_n_vectors(binary_data, n=20):
    if binary_data.shape[0] < n:
        raise IndexError("The dataset does not contain enough rows.")
    return binary_data.head(n)

def calculate_jaccard_and_smc(matrix):
    num_vectors = matrix.shape[0]
    jc_matrix = np.zeros((num_vectors, num_vectors))
    smc_matrix = np.zeros((num_vectors, num_vectors))

    for i in range(num_vectors):
        for j in range(i, num_vectors):
            vector1 = matrix[i]
            vector2 = matrix[j]

            f11 = np.sum((vector1 == 1) & (vector2 == 1))
            f01 = np.sum((vector1 == 0) & (vector2 == 1))
            f10 = np.sum((vector1 == 1) & (vector2 == 0))
            f00 = np.sum((vector1 == 0) & (vector2 == 0))

            jc = f11 / (f01 + f10 + f11) if (f01 + f10 + f11) != 0 else 0
            smc = (f11 + f00) / (f00 + f01 + f10 + f11) if (f00 + f01 + f10 + f11) != 0 else 0

            jc_matrix[i, j] = jc
            jc_matrix[j, i] = jc  # Symmetric matrix

            smc_matrix[i, j] = smc
            smc_matrix[j, i] = smc  # Symmetric matrix

    return jc_matrix, smc_matrix

def calculate_cosine_similarity(matrix):
    if matrix.shape[1] == 0:
        raise ValueError("No features to calculate cosine similarity.")
    return cosine_similarity(matrix)

def plot_heatmap(matrix, title, filename):
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
    plt.title(title)
    plt.savefig(filename)
    plt.show()

def main():
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'Purchase data'
    try:
        data = load_excel_data(file_path, sheet_name)
        binary_data = filter_binary_data(data)
        vectors = get_first_n_vectors(binary_data, 20)
        matrix = vectors.values

        if matrix.size == 0:
            raise ValueError("The matrix is empty after filtering.")

        jc_matrix, smc_matrix = calculate_jaccard_and_smc(matrix)
        cosine_sim_matrix = calculate_cosine_similarity(matrix)

        plot_heatmap(jc_matrix, 'Jaccard Coefficient Heatmap', 'jaccard_heatmap.png')
        plot_heatmap(smc_matrix, 'Simple Matching Coefficient Heatmap', 'smc_heatmap.png')
        plot_heatmap(cosine_sim_matrix, 'Cosine Similarity Heatmap', 'cosine_heatmap.png')

    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
